```python
def alokasi(M, N, UkuranElemen):
    """Mengalokasikan memori untuk matriks M x N. Setiap elemen matriks berukuran UkuranElemen byte."""
    larik = []
    for i in range(M):
        larik.append([0 for _ in range(N * UkuranElemen)])
    return larik

def xalloc(ukuran):
    """Mengalokasikan memori dan memeriksa apakah alokasi memori berhasil."""
    p = [0] * ukuran
    if p is None:
        print("Memori tidak cukup untuk alokasi matriks")
        exit(0)
    return p
```

Kode tersebut adalah implementasi dari dua fungsi yang berhubungan dengan alokasi memori untuk matriks.
1. **`alokasi(M, N, UkuranElemen)`**:
   - Tujuan: Fungsi ini bertujuan untuk mengalokasikan memori untuk matriks dengan ukuran M x N, di mana setiap elemen matriks memiliki ukuran `UkuranElemen` byte.
   - Cara Kerja: Fungsi ini menggunakan dua loop untuk membuat matriks dengan ukuran M x N. Setiap elemen dari matriks diinisialisasi dengan 0. Hasilnya adalah matriks `larik`.
   - Return Value: Fungsi mengembalikan matriks `larik` yang telah dialokasikan.

2. **`xalloc(ukuran)`**:
   - Tujuan: Fungsi ini bertujuan untuk mengalokasikan memori dengan ukuran tertentu dan memeriksa apakah alokasi memori berhasil.
   - Cara Kerja: Fungsi ini mencoba mengalokasikan memori untuk larik dengan panjang `ukuran`. Jika alokasi berhasil, larik dengan nilai awal 0 akan dibuat. Jika alokasi gagal, pesan kesalahan akan dicetak dan program akan keluar.
   - Return Value: Jika alokasi berhasil, fungsi mengembalikan larik yang telah dialokasikan.

```python
from PIL import Image

def tampilkan_citra(r, g, b, M, N):
    """Menampilkan citra yang berukuran M x N pixel ke layar."""
    width = N
    height = M
    image = Image.new("RGB", (width, height))

    for i in range(M):
        for j in range(N):
            set_pixel(image, j, i, r[i][j], g[i][j], b[i][j])

    image.show()

def set_pixel(image, i, j, r, g, b):
    """Menampilkan pixel dengan komponen rgb pada koordinat i, j."""
    image.putpixel((i, j), (r, g, b))

# Contoh penggunaan:
M = 480
N = 640

# Membuat citra R, G, B
r = [[255 for _ in range(N)] for _ in range(M)]
g = [[0 for _ in range(N)] for _ in range(M)]
b = [[0 for _ in range(N)] for _ in range(M)]

# Menampilkan citra
tampilkan_citra(r, g, b, M, N)
```

```python
def baca_citra_dari_arsip(nama_arsip):
    """Membaca citra dari file citra mentah. Citra hasil pembacaan disimpan di dalam matriks f."""
    try:
        with open(nama_arsip, "rb") as fp:
            M = int.from_bytes(fp.read(2), byteorder='little', signed=False)  # Baca tinggi citra
            N = int.from_bytes(fp.read(2), byteorder='little', signed=False)  # Baca lebar citra

            f = alokasi(M, N)  # Alokasi memori matriks untuk citra f

            if f is None:
                print("Memori tidak cukup")
                exit(0)

            # Baca data citra baris demi baris
            for i in range(M):
                # Baca data citra baris ke-i
                f[i] = list(fp.read(N))

            return f

    except FileNotFoundError:
        print("Arsip tidak ada")
        exit(0)

def alokasi(M, N):
    """Mengalokasikan memori untuk matriks M x N."""
    try:
        return [[0 for _ in range(N)] for _ in range(M)]
    except MemoryError:
        return None
nama_arsip = "nama_file.dat"  # Ganti dengan nama file dan ekstensi Anda
citra = baca_citra_dari_arsip(nama_arsip)
```

Jadi, keseluruhan kode ini berfungsi untuk membaca citra dari sebuah file citra mentah dan menyimpannya dalam matriks citra.

```python
def tulis_citra_ke_arsip(nama_arsip, f):
    """Menulis citra f ke dalam arsip nama_arsip."""
    try:
        M = len(f)
        N = len(f[0])

        with open(nama_arsip, "wb") as fp:
            fp.write(M.to_bytes(2, byteorder='little'))  # Tulis tinggi citra
            fp.write(N.to_bytes(2, byteorder='little'))  # Tulis lebar citra

            # Tulis data citra baris demi baris
            for i in range(M):
                # Tulis data citra baris ke-i
                fp.write(bytes(f[i]))

    except Exception as e:
        print(f"Terjadi kesalahan saat menulis arsip: {e}")

# Contoh penggunaan:
nama_arsip = "nama_file.dat"  # Ganti dengan nama file dan ekstensi Anda
citra = [[255, 0, 0], [0, 255, 0], [0, 0, 255]]  # Contoh citra
tulis_citra_ke_arsip(nama_arsip, citra)
```
